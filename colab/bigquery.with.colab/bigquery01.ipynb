```python
sample_count = 100
df = pd.io.gbq.read_gbq('''
  select name, sum(number) as count
  from `bigquery-public-data.usa_names.usa_1910_2013`
  where state = 'TX'
  group by name
  order by count desc
  limit 100
  ''', project_id = 'my-project-1011-1012', dialect='standard')

```
